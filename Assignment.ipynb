```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
data.head()
```

```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Low speed"
    elif val == speed_mean:
        return "neutral speed"
    else:
        return "High speed"
```

```python
# "High_Low_speed"  -> "High speed" , "Low speed"
data["High_Low_speed"] = data["Speed"].apply(set_speed)
data
```

```python
HP_mean = data["HP"].mean()

def set_HP(val):
    if val < HP_mean:
        return "Low HP"
    elif val == HP_mean:
        return "neutral HP"
    else:
        return "High HP"
```

```python
# "High_Low_HP"  -> "High HP" , "Low HP"
data["High_Low_HP"] = data["HP"].apply(set_HP)
data
```
